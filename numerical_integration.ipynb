```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz, simps, romberg
```

```python
def trapezoidal_rule(func, a, b, n):
    h = (b - a) / n
    integral = 0.5 * (func(a) + func(b))

    for i in range(1, n):
        x = a + i * h
        integral += func(x)

    integral *= h
    return integral

# Define the function to integrate
def f(x):
    #y = x**2/((x+1)**3)
    y = x**2/((x+1)**3)
    return y

# Integration limits and number of subintervals
a = 0
b = 100
n = 50

# Perform numerical integration using the trapezoidal rule
result = trapezoidal_rule(f, a, b, n)
print("Approximate value of the definite integral:", result)

# Generate x and y values for plotting the function
x_vals = np.linspace(a, b, 1000)
y_vals = f(x_vals)

# Plot the function
plt.figure(figsize=(18, 6))
plt.plot(x_vals, y_vals, label='sin(x)', color='blue')

# Plot the trapezoids for numerical integration
for i in range(n):
    x0 = a + i * (b - a) / n
    x1 = a + (i + 1) * (b - a) / n
    y0 = f(x0)
    y1 = f(x1)

    plt.fill([x0, x1, x1, x0], [0, 0, y1, y0], 'orange', edgecolor='black', alpha=0.5)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
#plt.xticks(np.arange(0,2*np.pi, step=5))
#plt.title('Numerical Integration Using Trapezoidal Rule')
#plt.legend()
#plt.grid(True)
plt.axhline(0,color='k')

# Show the plot
plt.savefig('./trapz_sin_blocks.jpg')
plt.show()
```

```python
# Define the function to integrate
def function(x):
    y = x**2/((x+1)**3)
    return y

# Integration limits and number of subintervals
a = 0
b = 2*np.pi
n = 100

# Generate x and y values for plotting the function
x_vals = np.linspace(a, b, n)
#print("x_vals",x_vals)

y_vals = function(x_vals)
#print("y_vals", y_vals)


# Perform numerical integration using SciPy's trapezoidal rule
integral = trapz(y_vals, x_vals)
print("Approximate value of the definite integral:", integral)

#---------------For plot------------------

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='sin(x)', color='blue')

# Plot the trapezoids for numerical integration
plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= b), color='orange', alpha=0.5)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Numerical Integration Using Trapezoidal Rule (SciPy)')
plt.legend()
#plt.grid(True)

# Show the plot
plt.show()
```
